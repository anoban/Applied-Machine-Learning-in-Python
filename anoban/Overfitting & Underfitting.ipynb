```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2023 - 5 - 4)
```

# ___Overfitting & Underfitting___
----------------

```python
# The definition of success of supervised machine learning models is their ability to make correct predictions on previously
# unseen data.
# If the performance is measured based on the predictions on training data, 
# ML models could simply memorize the inputs and make predictions.

# The real benefits of ML models come from their ability to generalize well to previously unseen data.
# This ability to make correct predictions on hold out datasets, is the ML model's ability to generalize.
```

```python
# A typical machine learning ML algorithm makes the following assumptions;
    # 1) Test sets / future datasets will have the same properties as the datasets used to train the model.
    # 2) Thus, the models that were accurate on the training set are expected to be accurate on the test set.
    # 3) But this may not happen sometimes due to the model being too tuned to the training dataset.
```

## ___Overfitting___

```python
# Models that are too complex (extremely fine tuned) to the training data are said to overfit and are less likely to generalize to
# new examples.
# Overfitting typically occurs when a complex model is fitted with inadequate amount of training data.
# An overfitted model excels at discerning very specific granular patterns in the training data but may fail to see the bigger picture and
# realize global patterns.
# Taking in the global patterns from the training datasets is critical for the models ability to generalize well.
```

### ___Overfitting in Regression___

```python
# a smooth curve
_x = np.square(np.arange(-10, 10, 0.1))

# add some noise
x = _x + np.random.randint(low = 5, high = 20, size = x.size)

xx = np.square(np.arange(-7, 3, 0.05))

plt.rcParams["figure.figsize"] = 10, 10
plt.plot(x, marker = "o", ms = 10, mfc = "yellow", mec = "black", linestyle = "none")
plt.plot(x, marker = "none", color = "red", label = "Overfitted")
plt.plot(_x + 10, marker = "none", color = "black", label = "Optimal fitting")
plt.plot(xx + 10, marker = "none", color = "blue", label = "Underfitted")
plt.legend()
plt.show()
```

```python
# This is a regression example.

# The yellow circles are thedata points in the training dataset.
```

```python
# The black line is the learned pattern of the optimal model.
# See, how it generalizes well over all the training set points.
```

```python
# The blue line represents the learned patterns of a vague model.
# It did not manage to learn the training data adequately to discern the patterns accurately. -> underfitted
# Thus, will produce inaccurate predictions on test datasets.
```

```python
# The red line shows the patterns learned by an overfitted model.
# It was too attuned to the training set by finetuning the model parameters.
# See how it manages to recognize the granular variation patterns between individual data points. -> overfitted
# The problem with this kind of models is that they won't generalize well to new input datasets,
# making far less accurate predictions.
```
