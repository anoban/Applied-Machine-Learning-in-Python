```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# ___K Nearest Neighbours (KNN) Classification___
--------------

```python
# KNN classifiers are instance/memory based classifiers.
# KNN algorithms can be used for classification & regression.

# Instance or memory based models work by memorizing the labelled examples they're fed with in the training dataset,
# and used this memory to classify new objects later.
```

```python
# The 'k' in KNN is the number of nearest neighbours the classifier takes into account in learning as well as making predictions.
# KNN algorithms in general have three steps.
```

```python
# Given a training set train_x with labels train_y and given a new instance of test_x to be classified,

# 1) Finds the most similar instances to test_x that are in train_x -> (call them xNN) ('x' nearest neighbours)
# 2) Find the labels of these xNN -> call them yNN. ('y' labels for the 'x' identified nearest neighbours)
# 3) Predict the label for test_x by combining the labels yNN -> majority vote.
```

```python
fruits = pd.read_table(r"../resources/assets/fruit_data_with_colors.txt", delimiter = "\t")
```

```python
# The original distribution.

fig, axes = plt.subplots()
fig.set_size_inches(8, 8)

for frt_name in fruits.fruit_name.unique():
    _fruit = fruits.query("fruit_name == @frt_name")
    axes.scatter(_fruit.height, _fruit.width, s = 100, marker = "o", edgecolor = "black", label = frt_name)
axes.set_xlabel("Height")
axes.set_ylabel("Width")
axes.legend()
plt.show()
```

## ___`k = 1`___

```python
fig, axes = plt.subplots()
fig.set_size_inches(8, 8)

for frt_name in fruits.fruit_name.unique():
    _fruit = fruits.query("fruit_name == @frt_name")
    axes.scatter(_fruit.height, _fruit.width, s = 100, marker = "o", edgecolor = "black", label = frt_name)
axes.set_xlabel("Height")
axes.set_ylabel("Width")
axes.legend()
axes.scatter(4, 9.4, c = "yellow", s = 200, edgecolor = "black", marker = "*")
axes.scatter(5, 6.5, c = "cyan", s = 100, edgecolor = "black", marker = "s")
plt.show()
```

```python
# Consider our yellow star shaped query point!
# What type of fruit could it be?

# What type of fruit the cyan box could be?
```

```python
# Cyan square has orange circles as the nearest neighbours. -> mandarins.
# Yellow star has the blue circles as the nearest neighbours. -> apples.
```

```python
# These results are from a KNN classification where N = 1.
# We are considering just the most closest neighbour for the classification.
```

```python
# If we kept on classifying like this for a large enough points, we will see a region that is dominated by a certain type of class.
# The margins of these regions will define the decision boundaries for the classifier.
# for example, if a data point falls inside a region dominated by apples, it is likely that it also is an apple.
```

```python
# To find the distance between two data points one could simply use the eculidean distance. (using Pythagorean theorem)
# Or one could use a weighted distance metric.
```

```python
# For example we can give the width and height the same weight in the distance function
# Or we could give more weight to the width over the height, since larger fruits are typically wider than the smaller ones instead of being
# taller.
# So, we can give the width 4 times more weight than the height when calculating the distances.
```

```python
# This change in the distance function will affect the alignment of the decission boundaries and hence,
# how the new points are classified by the KNN classifier.
```

## ___`n = ?`___

```python
fig, axes = plt.subplots()
fig.set_size_inches(8, 8)

for frt_name in fruits.fruit_name.unique():
    _fruit = fruits.query("fruit_name == @frt_name")
    axes.scatter(_fruit.height, _fruit.width, s = 100, marker = "o", edgecolor = "black", label = frt_name)
axes.set_xlabel("Height")
axes.set_ylabel("Width")
axes.legend()
axes.scatter(7.75, 7.74, c = "yellow", s = 200, edgecolor = "black", marker = "*")
axes.scatter(7.5, 6.5, c = "cyan", s = 100, edgecolor = "black", marker = "s")
plt.show()
```

```python
# kNN is sensitive to values in different scales.
# So, one must normalize the variables before fitting the model.
```

```python
# When k = 1, we'll get just one prediction from the most closest neighbour.
# When k > 1, we'll get more prediction from which we have to choose the majority vote.
```

```python
# If n = 2, the cyan square will be classified as orange.
# If n = 5, it will be classified as lemon

# If n = 2, the yellow star will be classified as orange.
# If n = 5, it will be classified as apple.
```

```python
# So, the predictions of kNN classifiers are sensitive to the number of neighbours we choose.
```
